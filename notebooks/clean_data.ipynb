### Data Cleaning

```python

#Rename columns
def rename_columns_(df):
    # Dictionnaire de correspondance des anciens et nouveaux noms de colonnes
    column_mapping = {
        'CRIM': 'PerCapCrimeRate',
        'ZN': 'ResidentialLandOver25K',
        'INDUS': 'NonRetailAcresRatio',
        'CHAS': 'CharlesRiverDummy',
        'NOX': 'NitricOxideConc',
        'RM': 'AvgNumRooms',
        'AGE': 'Pre1940sOwnerOcc',
        'DIS': 'WeightedDistToEmploy',
        'RAD': 'RadialHighwayAccessIndex',
        'TAX': 'MedianHomeTaxRate',
        'PTRATIO': 'PupilTeacherRatio',
        'B': 'AfroAmericanProportion',
        'LSTAT': 'LowerStatusRatio',
        'MEDV': 'MedianHomeValue'
    }
    
    # Renommer les colonnes dans le DataFrame
    df.rename(columns=column_mapping, inplace=True)
    
    return df
```

```python
df = rename_columns_(df)
```
